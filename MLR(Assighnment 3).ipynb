#### The Cylinder column is ignored because it contains only a single repeated value throughout the dataset, meaning it has no variation.
#### Since there is no variability, this feature does not contribute any useful information for analysis or prediction.
#### In visualizations, such as a heatmap, it does not show meaningful correlation with other variables.
#### Similarly, in a scatter plot, it appears as a straight line, indicating no relationship or distribution pattern.

#### Core concept:
#### A variable with constant value = zero variance = no predictive power

#### Columns like Fuel Type, Automatic, Doors, and Gears have a small number of unique values (e.g., 2, 3, or 4), so they behave like categorical/discrete variables.
#### For such variables, a countplot is more appropriate because it shows the frequency distribution of each category clearly.
#### A boxplot is not very meaningful here, since these variables do not have enough continuous variation to properly display median, quartiles, and spread.

#### important:-Boxplot can technically be used on numeric discrete data, but with very few unique values, it becomes uninformative or misleading, not wrong.
#### Few unique values → treat as categorical → use countplot
#### Continuous values → use boxplot

## Conclusion

### Price is considered as the target variable, and its relationship with other variables is analyzed using the heatmap.

### From the analysis, Age_08_04 shows a strong negative relationship with Price (-0.88), indicating that as the age of the car increases, the price decreases significantly. Similarly, KM also has a negative relationship with Price(-0.57), meaning that cars with higher mileage tend to have lower prices.

### On the other hand, Weight has a strong positive relationship with Price (0.58), suggesting that heavier cars generally have higher prices. HP (Horsepower) also shows a positive relationship with Price (0.31), indicating that better performance contributes to higher value.

### The heatmap also highlights the presence of multicollinearity among some independent variables. A very strong multicollinearity is observed between Fuel_Type_Diesel and Fuel_Type_Petrol (-0.94), which is due to the use of dummy variables (one-hot encoding). Therefore, one of these variables should be removed to avoid the dummy variable trap.

### Additionally, moderate multicollinearity is observed between Weight and Fuel Type variables(0.57, -0.56), as well as between KM and Age(0.51), since older cars tend to have higher mileage.

## Overall, the analysis concludes that Age, KM, Weight, and HP are the most important factors influencing Price, while multicollinearity needs to be handled carefully for accurate model building

### Check multicollinearity in features

```python
import statsmodels.formula.api as smf
```

```python

model=smf.ols("Age_08_04 ~ KM",data=df).fit()
print(model.summary())

model.fittedvalues

model.resid

MSE=np.mean(model.resid**2)
print(MSE)

r2=model.rsquared
print(r2)

vif=1/(1-r2)
print("VIF Of age and KM:",vif)
```

### There is no multicollinearity issue in between age and km cause vif value is 1.34 which is safe from multicollinearity issue
#### ====================================================================

```python

model=smf.ols("Price ~ Age_08_04",data=df).fit()
print(model.summary())

model.fittedvalues

model.resid

MSE=np.mean(model.resid**2)
print(MSE)

r2=model.rsquared
print(r2)

vif=1/(1-r2)
print("VIF Of price and age:",vif)
```

### There is no multicollinearity issue in age and price which is good

### perform lasso regression

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=7.0)  
lasso.fit(x_train, y_train)
y_pred_train=lasso.predict(x_train)
y_pred_test=lasso.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error
train_error=np.sqrt(mse(y_train,y_pred_train))

test_error=np.sqrt(mse(y_test,y_pred_test))

print("Train error:",train_error)
print("Test error:",test_error)
```

```python
X=df[['Age_08_04','KM','HP','Automatic','cc','Doors','Gears','Weight']]
d1=pd.DataFrame(list(X))
d2=pd.DataFrame(lasso.coef_)

df=pd.concat([d1,d2],axis=1)
df.columns=["Feature","coefficient"]
df
```
